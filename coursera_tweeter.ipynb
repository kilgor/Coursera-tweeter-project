```python
# Get the counts of True (retweet) and False (original tweet)
retweet_count = df['is_retweet'].value_counts()

# Print the number of retweets (True) and original tweets (False)
print('Number of Retweets:', retweet_count.get(True, 0))
print('Number of Original Tweets:', retweet_count.get(False, 0))
```

```python
# Sum the 'user_favourites' column to get the total number of favorites (likes)
total_favorites = df['user_favourites'].sum()
print('Total Number of Favorites:', total_favorites)
```

```python
#first drop is_retweet column and check for data again
df.drop('is_retweet', axis=1, inplace=True)
df.head()
```

```python
#lets check dated and in which month the tweets were made in and make a bar chart
#and convert the numbers of months to names of months
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
month_tweets = df['month'].value_counts()
print(month_tweets)

plt.figure(figsize=(10,10))
sns.barplot(x=month_tweets.values, y=month_tweets.index, palette='viridis')
plt.title('Number of Tweets by Month')
plt.xlabel('Number of Tweets')
plt.ylabel('Month')
plt.show()
```

```python
#lets check the number of tweets made by verified users and unverified users
verified_tweets = df[df['user_verified'] == True]['user_verified'].count()
unverified_tweets = df[df['user_verified'] == False]['user_verified'].count()
print('Number of Tweets by Verified Users:', verified_tweets)

print('Number of Tweets by Unverified Users:', unverified_tweets)

#and make a pie chart
plt.figure(figsize=(7,7))
plt.pie([verified_tweets, unverified_tweets], labels = ['Verified', 'Not Verified'], autopct='%1.1f%%', startangle=90)
plt.title('Verified vs Unverified Users')
plt.show()
```
